```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('Placement_Data_Full_Class.csv')
```

```python
import pandas as pd
import numpy as np
from pandas.api.types import CategoricalDtype
from plotnine import *
from plotnine.data import mpg
%matplotlib inline

(ggplot(dataset)
 + aes(x='gender')
 + geom_bar(size=20)
)
```

```python
(ggplot(dataset)
 + aes(x='gender')
 + geom_bar(size=20)
 + coord_flip()        # flipping the x- and y-axes
 + labs(title='Number of Male and Female', x='Gender', y='count') # customizing labels
)
```

```python

(ggplot(dataset)
 + aes(x='gender',y='ssc_p',fill='ssc_b')
 + geom_boxplot(size=20)
 + labs(title='Number of Male and Female', x='Gender', y='count')
)
```

```python
(ggplot(dataset, aes(x='specialisation', y='salary', fill='specialisation')) +
        geom_boxplot()
)
```
