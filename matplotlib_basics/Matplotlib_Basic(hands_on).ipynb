```python
import matplotlib.pyplot as plt
import numpy as np
```

## Simple Plot

```python
plt.plot([1, 2, 3, 4], [1, 4, 2, 3])
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Plot')
plt.show()
```

### Plot with labels

```python
plt.plot([1, 2, 3, 4], [1, 4, 2, 3], marker='o', color='blue', linestyle='--')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude (mV)')
plt.title('Data over Time')
plt.grid(True)
plt.show()
```

## Subplot

```python
fig, axs = plt.subplots(2, 2)
axs[0, 0].plot([1, 2, 3], [1, 2, 3])
axs[0, 0].set_title('Subplot 1')
axs[0, 1].plot([1, 2, 3], [3, 2, 1], color='red')
axs[0, 1].set_title('Subplot 2')
axs[1, 0].plot([1, 2, 3], [1, 3, 2], color='green')
axs[1, 0].set_title('Subplot 3')
axs[1, 1].plot([1, 2, 3], [2, 1, 3], color='purple')
axs[1, 1].set_title('Subplot 4')
plt.tight_layout()
plt.show()
```

### Advanced plot

```python
fig, ax = plt.subplots()
ax.plot([1, 2, 3, 4], [1, 4, 2, 3], marker='o', linestyle='--', color='blue', markersize=8, linewidth=2)
ax.set_title('Advanced Plot Example', fontsize=16, color='darkgreen')
ax.set_xlabel('X-axis Label', fontsize=12, color='darkred')
ax.set_ylabel('Y-axis Label', fontsize=12, color='darkblue')
ax.set_xlim(0, 5)
ax.set_ylim(0, 5)
ax.grid(True, linestyle=':', alpha=0.7)
ax.spines['bottom'].set_color('blue')
ax.spines['left'].set_color('red')
ax.spines['left'].set_linewidth(2)
ax.spines['right'].set_color('green')
ax.spines['top'].set_color('purple')
ax.spines['right'].set_linewidth(1)
ax.spines['top'].set_linewidth(1)
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(2, 2, figsize=(10, 8))


ax[0, 0].plot([1, 2, 3, 4], [1, 4, 2, 3], marker='o', color='blue')
ax[0, 0].set_title('Subplot 1: Growing Trend')
ax[0, 0].set_xlabel('X1-axis')
ax[0, 0].set_ylabel('Y1-axis')
ax[0, 0].grid(True)


ax[0, 1].plot([1, 2, 3, 4], [4, 2, 3, 1], marker='x', color='red', linestyle='--')
ax[0, 1].set_title('Subplot 2: Decreasing Trend')
ax[0, 1].set_xlabel('X2-axis')
ax[0, 1].set_ylabel('Y2-axis')
ax[0, 1].grid(True)


ax[1, 0].plot([1, 2, 3, 4], [2, 1, 4, 3], marker='s', color='green', linestyle=':')
ax[1, 0].set_title('Subplot 3: Volatile Data')
ax[1, 0].set_xlabel('X3-axis')
ax[1, 0].set_ylabel('Y3-axis')
ax[1, 0].grid(True)


ax[1, 1].plot([1, 2, 3, 4], [3, 4, 1, 2], marker='^', color='purple', linestyle='-.')
ax[1, 1].set_title('Subplot 4: Mixed Data')
ax[1, 1].set_xlabel('X4-axis')
ax[1, 1].set_ylabel('Y4-axis')
ax[1, 1].grid(True)

plt.suptitle('Advanced 2x2 Subplots Example', fontsize=16)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

# Bar chart

```python
langs = ['C', 'C++', 'Java', 'Python', 'PHP']
students = [23,17,35,29,12]
plt.bar(langs, students)
plt.xlabel('Programming Language')
plt.ylabel('Number of Students')
plt.title('Number of Students per Language')
plt.show()
```

```python
data = [[30, 25, 50, 20],
[40, 23, 51, 17],
[35, 22, 45, 19]]
X = np.arange(4)
print(X)
plt.bar(X + 0.00,data[0], color = 'b', width = 0.25, label='Group 1')
plt.bar(X + 0.25, data[1], color = 'g', width = 0.25, label='Group 2')
plt.bar(X + 0.50, data[2], color = 'r', width = 0.25, label='Group 3')
plt.grid(True)
plt.title("Grouped Bar Plot")
plt.xlabel("Category")
plt.ylabel("Values")
plt.legend()
plt.show()
```

## Histogram

```python
a = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27])
plt.hist(a, bins=100)
plt.title("histogram of result")
plt.xlabel('marks')
plt.ylabel('no. of students')
plt.show()
```

## Pie Chart

```python
plt.axis('equal')
langs = ['C', 'C++', 'Java', 'Python', 'PHP']
students = [23,17,35,29,12]
plt.pie(students, labels = langs,autopct='%0.2f%%')
plt.title('Distribution of Students by Language')
plt.show()
```

## Mesh Grid

```python
xlist = np.linspace(-3.0, 3.0, 100)
ylist = np.linspace(-3.0, 3.0, 100)
X, Y = np.meshgrid(xlist, ylist)
Z = np.sqrt(X**2 + Y**2)
cp=plt.contourf(X, Y, Z)
plt.grid(True)
plt.grid(color='b', ls = '-.', lw = 1)
plt.title("mesh plot",loc="right")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.colorbar(cp)
plt.show()
```

```python
from mpl_toolkits import mplot3d
fig = plt.figure()
ax = plt.axes(projection='3d')
xlist = np.linspace(-3.0, 3.0, 100)
ylist = np.linspace(-3.0, 3.0, 100)
X, Y = np.meshgrid(xlist, ylist)
Z = np.sqrt(X**2 + Y**2)
cp=plt.contourf(X, Y, Z)
ax.contour3D(X, Y, Z, 50, cmap='gray')
ax.plot_wireframe(X, Y, Z, color='black')
ax.plot_surface(X,Y,Z,cmap='viridis', edgecolor='none')
plt.show()
```

# Contour

```python
xlist =[4,2]
ylist = [4,2]
X, Y = np.meshgrid(xlist, ylist)

Z=X*Y
cp=plt.contourf(X, Y,Z)
plt.grid(True)
plt.grid(color='b', ls = '-.', lw = 1)
plt.title("Contour plot",loc="right")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.colorbar(cp)
plt.show()
```

## Gradient

```python
x,y = np.meshgrid(np.arange(-2, 2, .2), np.arange(-2, 2, .25))
z = x*np.exp(-x**2 - y**2)
v, u = np.gradient(z, .2, .2)
plt.subplots()
plt.quiver(x,y,u,v)
plt.show()
```

# Box plot

```python
np.random.seed(10)
collectn_1 = np.random.normal(100, 10, 200)
collectn_2 = np.random.normal(80, 30, 200)
collectn_3 = np.random.normal(90, 20, 200)
collectn_4 = np.random.normal(70, 25, 200)
data_to_plot=[collectn_1,collectn_2,collectn_3,collectn_4]
plt.boxplot(data_to_plot)
plt.title('Box Plot of Collections')
plt.xlabel('Collection Group')
plt.ylabel('Values')
plt.show()
```

## Violin plot

```python
np.random.seed(10)
collectn_1 = np.random.normal(100, 10, 200)
collectn_2 = np.random.normal(80, 30, 200)
collectn_3 = np.random.normal(90, 20, 200)
collectn_4 = np.random.normal(70, 25, 200)
data_to_plot=[collectn_1,collectn_2,collectn_3,collectn_4]
plt.violinplot(data_to_plot)
plt.title('Violin Plot of Collections')
plt.xlabel('Collection Group')
plt.ylabel('Values')
plt.show()
```

# 3D

```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
fig = plt.figure()
ax = plt.axes(projection='3d')
z = np.linspace(0, 1, 100)
x = z * np.sin(20 * z)
y = z * np.cos(20 * z)
ax.plot3D(x, y, z, 'gray')
ax.set_title('3D Line Plot')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
plt.show()
```
